```python
import json
import pandas as pd

# Load the JSON data
with open("user-wallet-transactions.json", "r") as file:
    data = json.load(file)

# Convert to a DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
df['actionData'] = pd.to_numeric(df['actionData'], errors='coerce')
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
# Group transactions by wallet
group = df.groupby('userWallet')

# Feature engineering
features = pd.DataFrame()

features['num_transactions'] = group.size()
features['num_borrows'] = group['action'].apply(lambda x: (x == 'borrow').sum())
features['num_repays'] = group['action'].apply(lambda x: (x == 'repay').sum())
features['num_deposits'] = group['action'].apply(lambda x: (x == 'deposit').sum())
features['num_liquidations'] = group['action'].apply(lambda x: (x == 'liquidationcall').sum())
features['num_redeems'] = group['action'].apply(lambda x: (x == 'redeemunderlying').sum())

features['total_borrowed'] = group.apply(lambda x: x.loc[x['action'] == 'borrow', 'actionData'].sum())
features['total_repaid'] = group.apply(lambda x: x.loc[x['action'] == 'repay', 'actionData'].sum())
features['repay_ratio'] = features['total_repaid'] / features['total_borrowed']
features['repay_ratio'].fillna(0, inplace=True)

features = features.fillna(0)
features.head()
```

```python
# Score from 0 to 1000
def score_wallet(row):
    score = 0

    # Penalize liquidations
    score -= row['num_liquidations'] * 100

    # Reward responsible borrowing/repayment
    score += min(row['repay_ratio'], 1.0) * 500

    # Reward activity (deposits, redemptions, etc.)
    score += min(row['num_deposits'] + row['num_redeems'], 10) * 20

    # Clip between 0 and 1000
    return max(0, min(1000, score))

features['credit_score'] = features.apply(score_wallet, axis=1)
features[['credit_score']].head(10)
```

```python
features[['credit_score']].to_csv("wallet_credit_scores.csv")
```
